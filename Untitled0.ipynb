```python
def get_user_input():
    """Запрашивает у пользователя цены, скидку и порог элитности"""
    prices_input = input("Введите цены товаров (через пробел): ")
    prices = [float(x) for x in prices_input.split()]

    discount_percent = float(input("Введите процент скидки (например, 10): "))
    threshold = float(input("Введите порог элитного товара (например, 1000): "))

    return prices, discount_percent, threshold


def calculate_statistics(prices):
    """Вычисляет статистику по ценам"""
    total = sum(prices)
    count = len(prices)
    avg = total / count if count > 0 else 0
    min_price = min(prices) if prices else 0
    max_price = max(prices) if prices else 0

    return {
        "count": count,
        "total": total,
        "min": min_price,
        "max": max_price,
        "avg": avg
    }


def calculate_discount(total_sum, discount_percent):
    """Вычисляет скидку и итоговую сумму"""
    discount_amount = total_sum * (discount_percent / 100)
    final_sum = total_sum - discount_amount

    return discount_amount, final_sum


def filter_elite_items(prices, threshold):
    """Фильтрует товары дороже порога"""
    elite_items = [price for price in prices if price > threshold]
    return elite_items


def print_receipt(stats_dict, discount_data, elite_items):
    """Выводит красивый чек с визуализацией"""
    # Распаковываем данные
    count = stats_dict["count"]
    total = stats_dict["total"]
    min_price = stats_dict["min"]
    max_price = stats_dict["max"]
    avg = stats_dict["avg"]

    discount_amount, final_sum = discount_data

    print("\n---------- ВАШ ЧЕК ----------")
    print(f"Количество позиций: {count}")
    print(f"Сумма без скидки: {total:.2f} руб.")
    print(f"Скидка: -{discount_amount:.2f} руб.")
    print(f"ИТОГО К ОПЛАТЕ: {final_sum:.2f} руб.")
    print("-----------------------------")
    print(f"Самый дешевый: {min_price:.2f} руб.")
    print(f"Самый дорогой: {max_price:.2f} руб.")
    print(f"Средний чек: {avg:.2f} руб.")

    # Выводим элитные товары
    print("\nСписок элитных товаров (дороже порога):")
    if elite_items:
        for price in elite_items:
            print(f"- {price:.2f} руб.")
    else:
        print("Нет элитных товаров")

    # Визуализация цен звездочками
    print("\nГрафик цен (визуализация):")
    if count > 0:
        max_price_for_scale = max_price
        for i, price in enumerate(stats_dict.get("prices_original", []), 1):
            # Масштабируем: максимум = 30 звездочек
            stars_count = int((price / max_price_for_scale) * 10) if max_price_for_scale > 0 else 0
            print(f"Товар {i}: {'*' * stars_count}")
    print("-----------------------------")


def main():
    """Главная функция-диспетчер"""
    # 1. Получаем данные от пользователя
    prices, discount_percent, threshold = get_user_input()

    # Сохраняем оригинальные цены для графика
    original_prices = prices.copy()

    # 2. Вычисляем статистику
    stats_dict = calculate_statistics(prices)

    # Добавляем оригинальные цены в словарь для графика
    stats_dict["prices_original"] = original_prices

    # 3. Вычисляем скидку
    total_sum = stats_dict["total"]
    discount_data = calculate_discount(total_sum, discount_percent)

    # 4. Фильтруем элитные товары
    elite_items = filter_elite_items(prices, threshold)

    # 5. Печатаем чек
    print_receipt(stats_dict, discount_data, elite_items)


# Запуск программы
if __name__ == "__main__":
    main()
```

```python
def get_user_input():
    """Запрашивает у пользователя цены, скидку и порог элитности"""
    prices_input = input("Введите цены товаров (через пробел): ")
    prices = [float(x) for x in prices_input.split()]

    discount_percent = float(input("Введите процент скидки (например, 10): "))
    threshold = float(input("Введите порог элитного товара (например, 1000): "))

    return prices, discount_percent, threshold


def calculate_statistics(prices):
    """Вычисляет статистику по ценам"""
    total = sum(prices)
    count = len(prices)
    avg = total / count if count > 0 else 0
    min_price = min(prices) if prices else 0
    max_price = max(prices) if prices else 0

    # Возвращаем КОРТЕЖ (просто перечисляем значения)
    return count, total, min_price, max_price, avg


def calculate_discount(total_sum, discount_percent):
    """Вычисляет скидку и итоговую сумму"""
    discount_amount = total_sum * (discount_percent / 100)
    final_sum = total_sum - discount_amount

    return discount_amount, final_sum


def filter_elite_items(prices, threshold):
    """Фильтрует товары дороже порога"""
    elite_items = [price for price in prices if price > threshold]
    return elite_items


def print_receipt(count, total, min_price, max_price, avg, discount_data, elite_items, original_prices):
    """Выводит красивый чек с визуализацией"""

    discount_amount, final_sum = discount_data

    print("\n---------- ВАШ ЧЕК ----------")
    print(f"Количество позиций: {count}")
    print(f"Сумма без скидки: {total:.2f} руб.")
    print(f"Скидка: -{discount_amount:.2f} руб.")
    print(f"ИТОГО К ОПЛАТЕ: {final_sum:.2f} руб.")
    print("-----------------------------")
    print(f"Самый дешевый: {min_price:.2f} руб.")
    print(f"Самый дорогой: {max_price:.2f} руб.")
    print(f"Средний чек: {avg:.2f} руб.")

    # Выводим элитные товары
    print("\nСписок элитных товаров (дороже порога):")
    if elite_items:
        for price in elite_items:
            print(f"- {price:.2f} руб.")
    else:
        print("Нет элитных товаров")

    # Визуализация цен звездочками
    print("\nГрафик цен (визуализация):")
    if count > 0:
        max_price_for_scale = max_price
        for i in range(len(original_prices)):
            price = original_prices[i]
            stars_count = int((price / max_price_for_scale) * 10) if max_price_for_scale > 0 else 0
            print(f"Товар {i + 1}: {'*' * stars_count}")
    print("-----------------------------")


def main():
    """Главная функция-диспетчер"""
    # 1. Получаем данные от пользователя
    prices, discount_percent, threshold = get_user_input()

    # Сохраняем оригинальные цены для графика
    original_prices = prices.copy()

    # 2. Вычисляем статистику (получаем КОРТЕЖ из 5 чисел)
    count, total, min_price, max_price, avg = calculate_statistics(prices)

    # 3. Вычисляем скидку
    discount_data = calculate_discount(total, discount_percent)

    # 4. Фильтруем элитные товары
    elite_items = filter_elite_items(prices, threshold)

    # 5. Печатаем чек (передаём все переменные отдельно)
    print_receipt(count, total, min_price, max_price, avg, discount_data, elite_items, original_prices)


# Запуск программы
if __name__ == "__main__":
    main()
```
